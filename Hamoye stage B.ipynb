```python
# libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv("energydata_complete.csv") 
```

```python
feature=data[["T2"]]
target=np.array(data["T6"])
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score,mean_squared_error
```

```python
scale=MinMaxScaler()
feature=scale.fit_transform(feature)
```

```python
target.shape
```

```python
model=LinearRegression()
model.fit(feature,target)
```

```python
y_pred=model.predict(feature)
```

```python
# number 12

r2_scores=r2_score(target, y_pred)
r2_scores
```

```python
#preparing new data

new_data=pd.read_csv("energydata_complete.csv")
new_data=data.drop(["date", "lights"],axis=1)
scaler=MinMaxScaler()
new_data=pd.DataFrame(scaler.fit_transform(new_data),columns=new_data.columns)
new_x=new_data.drop("Appliances",axis=1)
new_y=new_data["Appliances"]
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y=train_test_split(new_x,new_y,random_state=42,test_size=0.3)
```

```python
train_x
```

```python
test_x
```

```python
new_model=LinearRegression()

new_model.fit(train_x,train_y)

y_pred_new=new_model.predict(test_x)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
#number 13

round(mean_absolute_error(y_pred_new,test_y),2)
```

```python
# number 14

round(mean_squared_error(y_pred_new,test_y),2)
```

```python
# number 15

rmse_of_lin_reg=np.sqrt(round(mean_squared_error(y_pred_new,test_y),2))
rmse_of_lin_reg
```

```python
#number 16(Coefficient of Determination )


r2=r2_score(test_y,y_pred_new)
r2
```

```python
#number 17(lowest and highest column)
index=list(range(len(new_model.coef_)))
zipped_form=list(zip(index,new_model.coef_))

zipped_form=sorted(zipped_form, key=lambda x: x[-1])

columns=train_x.columns

lowest,highest=train_x.columns[zipped_form[0][0]],train_x.columns[zipped_form[-1][0]]

lowest,highest
```

```python
#number 18

ridge=Ridge(0.4)
ridge.fit(train_x,train_y)
pred_ridge=ridge.predict(test_x)
rmse_of_ridge=np.sqrt(round(mean_squared_error(pred_ridge,test_y),2))
rmse_of_ridge==rmse_of_lin_reg
```

```python
#number 19(number of non zero)
lasso=Lasso(0.001)
lasso.fit(train_x,train_y)
pred_lasso=lasso.predict(test_x)

number_of_nun_zero=np.sum(lasso.coef_!=0)

number_of_nun_zero
```

```python
#number 20(lasso regression of rmse)

lasso_rsme=np.sqrt(round(mean_squared_error(pred_lasso,test_y),3))
lasso_rsme
```
